## Split the Data into Training and Testing Sets

```python
from pandas.core.frame import DataFrame
# Create the labels set `y` and features DataFrame `X`
X = data.copy()
X = X.drop(columns="spam")
X.head()
y = pd.Series(data['spam'])
y.head()
```

```python
# Check the balance of the labels variable (`y`) by using the `value_counts` function.
y.value_counts()
```

```python
# Split the data into X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X,y)
X_train.head()
```

## Scale the Features

Use the `StandardScaler` to scale the features data. Remember that only `X_train` and `X_test` DataFrames should be scaled.

```python
from sklearn.preprocessing import StandardScaler

# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_train_scaled = X_train.copy()
columns_to_scale = ["capital_run_length_average", "capital_run_length_longest", "capital_run_length_total"]
X_train_scaled = X_train_scaled.drop(columns = ['capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total'])
X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])
X_train_scaled.head()
```

```python
from pickle import TRUE
# Scale the training data
X_test_scaled = X_test.copy()
columns_to_scale = ["capital_run_length_average", "capital_run_length_longest", "capital_run_length_total"]
X_test_scaled = X_test_scaled.drop(columns = ['capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total'])
X_test_scaled[columns_to_scale] = scaler.fit_transform(X_test[columns_to_scale])
X_test_scaled.head()
```

## Create and Fit a Logistic Regression Model

Create a Logistic Regression model, fit it to the training data, make predictions with the testing data, and print the model's accuracy score. You may choose any starting settings you like.

```python
# Train a Logistic Regression model and print the model score
from sklearn.linear_model import LogisticRegression

logistic_regression_model = LogisticRegression(random_state=1)
logistic_regression_model.fit(X_train_scaled, y_train)

print(f"Training Data Score for Logistical Regression: {logistic_regression_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score Logistical Regression: {logistic_regression_model.score(X_test_scaled, y_test)}")

```

```python
# Make and save testing predictions with the saved logistic regression model using the test data
predictions = logistic_regression_model.predict(X_test_scaled)
# Review the predictions

df_results = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
df_results.head()
```

```python
# Calculate the accuracy score by evaluating `y_test` vs. `testing_predictions`.
print(f"The accuracy score of y_test predictions: {accuracy_score(y_test, predictions)}")
```

## Create and Fit a Random Forest Classifier Model

Create a Random Forest Classifier model, fit it to the training data, make predictions with the testing data, and print the model's accuracy score. You may choose any starting settings you like.

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=128, random_state=78)
rf_model_fit = rf_model.fit(X_train_scaled, y_train)

print(f"Training Data Score for Random Forest: {rf_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score Random Forest: {rf_model.score(X_test_scaled, y_test)}")
```

```python
# Make and save testing predictions with the saved logistic regression model using the test data
testing_predictions = rf_model.predict(X_test_scaled)
testing_predictions[:5]
df_testing_predictions = pd.DataFrame({"Prediction": testing_predictions, "Actual": y_test})
df_testing_predictions.head()
# Review the predictions
```

```python
# Calculate the accuracy score by evaluating `y_test` vs. `testing_predictions`.
print(f"The accuracy score of y_test predictions of Random Forest: {accuracy_score(y_test, testing_predictions)}")
```
